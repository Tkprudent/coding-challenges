```python
import csv
from pprint import pprint
csvFile = open("Clarksville_weather_history.csv", newline ='')
reader = csv.DictReader(csvFile)
myReader = []
for row in reader:
    if len(row['TAVG']) > 0:
        row['TAVG'] = float(row['TAVG'])
    else:
        row['TAVG'] = None 
    if len(row['TMAX']) > 0:
        row['TMAX'] = float(row['TMAX'])
    else:
        row['TMAX'] = None 
    if len(row['TMIN']) > 0:
        row['TMIN'] = float(row['TMIN'])
    else:
        row['TMIN'] = None 
    if len(row['PRCP']) > 0:
        row['PRCP'] = float(row['PRCP'])
    else:
        row['PRCP'] = None 
    if len(row['SNOW']) > 0:
        row['SNOW'] = float(row['SNOW'])
    else:
        row['SNOW'] = None 
    myReader.append(row)
```

2. How many unique weather station ids (STATION fields) are in the data file? Provide the count and then list them

4. List date range (DATE) for each station (STATION). Stations come and go, so their data is only available for limited date ranges.For each station, list the date of the earliest record in the file and the date of the last record in the file.

```python
def computeDate(dictt):
    unique_data = {}
    for row in dictt: 
        if row['STATION'] in unique_data:
            if row['DATE'] < unique_data[row['STATION']][0]:    
                unique_data[row['STATION']][0] = row['DATE']
            if row['DATE'] > unique_data[row['STATION']][1]:     
                unique_data[row['STATION']][1] = row['DATE']
        else:
            unique_data[row['STATION']] = [row['DATE'], row['DATE']]
    for key in unique_data:
        print("For station ", key, " the start date is ", unique_data[key][0], " and the end date is ", unique_data[key][1])

computeDate(myReader)     
```

5. List the station id with the highest elevation and the station with the lowest elevation.

```python
# Find the highest elevation
highestElevation = myReader[0]
for row in myReader:
    if row['ELEVATION'] > highestElevation['ELEVATION']:
        highestElevation = row

# Find the lowest elevation
lowestElevation = myReader[0]
for row in myReader:
    if row['ELEVATION'] < lowestElevation['ELEVATION']:
        lowestElevation = row
# output the stationID with the highest elevation
print("The stationID with the highest elevation is, {}, with an elevation {} ".format(highestElevation['STATION'],
            highestElevation['ELEVATION']))
# output the station with the lowest elevation
print("The station with the lowest elevation is, {}, with an elevation {} ".format(lowestElevation['NAME'],
            lowestElevation['ELEVATION']))
```
