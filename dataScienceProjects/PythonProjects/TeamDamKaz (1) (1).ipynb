###  Problem 1	

__What are the checkout and return kiosks with highest usage rate?__

Here we are interested in knowing the checkout and return kiosk with the highest usage, this is to possibly answer for the need of ensuring more provision of bicycles in such area so as to make sure that we give maximum satisfaction to our customers.

We used the count of the users that checkout in this location in the data collected as the means to get the solution to this problem.

```python
# This is the solution to the Checkout kiosk
checkout = dict(data['CheckoutKioskName'])
returnKiosk = dict(data['ReturnKioskName'])
checkoutCounts  = {}
for checkoutKiosk in checkout.values():
    checkoutCounts[checkoutKiosk] = checkoutCounts.get(checkoutKiosk,0)+1
from operator import itemgetter
topCheckoutpts = sorted(checkoutCounts.items(), key = itemgetter(1), reverse = True)
topFivepts = pd.DataFrame(topCheckoutpts[:5], columns = ['CheckoutKiosk', 'UsageRate'])
topFivepts
```

```python
topFivepts.plot(x = topFivepts['CheckoutKiosk'],kind = "barh", rot = 0, title = "Top 5 checkout points" )
plt.show()
```

We observed from the above solution the top five checkout points in our data, and we see that Shelby Bottoms Nature Center-1900 Davidson Str has the highest checkpout rate.

```python
rkioskCounts = {}
for rkiosk in returnKiosk.values():
    rkioskCounts[rkiosk] = rkioskCounts.get(rkiosk,0)+1
from operator import itemgetter
returnKioskpts = sorted(rkioskCounts.items(), key = itemgetter(1), reverse = True)
topFivereturnKiosk = pd.DataFrame(returnKioskpts[:5], columns = ['ReturnKiosk', 'Rate'])
topFivereturnKiosk 
```

```python
topFivepts.plot(x = topFivereturnKiosk['ReturnKiosk'],kind = "barh", rot = 0, title = "Top 5 Return Kiosk")
plt.show()
```

#### Output Explanation
Its interesting to note that we also have the same kiosk name reappearing as the highest return kiosk, this makes sense because we know for the most part that most people would return a bicycle to the exact location where they took it from, but we observe that the checkout and return do not have the same value, this is because there are cases where people take a bicycle from a point A and return it to another point. 

A case scenario might be in a place like Nashville, the Music City where we have several tourist attraction, a tourist might park his vehicle in a different location and choose to ride a bicycle to the  closest location to his parking lot.

### Problem 5 

 __What day of the week has the highest pickup rate?__

```python
import datetime
checkoutdate = data['CheckoutDateLocal']
#day_week = [df[date_column].dt.weekday_name]
checkout = list(checkoutdate)
day_week =[datetime.datetime.strptime(x, '%m/%d/%y').strftime('%A') for x in checkout]
days = set(day_week)
for day in days:
    cnt = day_week.count(day)
    print("{} appears {} times".format(day,cnt))
```

```python
df = pd.DataFrame(day_week, columns = ["weekDay"] )
dfCnt = pd.value_counts(df['weekDay'].values, sort=False)
dfCnt.plot(kind = "bar", title = "WeekDay with Highest Pickup Rate")
plt.show()
```

#### Result Explanation
From the output and the plot, we can observe that Saturday has the highest pickup rate followed by Sunday
and then Friday. This is telling us that people go out for bike riding on weekends which is realistic.

### Problem 6

__What season of the year has the highest trip duration?__

We are interested in knowing what season of the year has the highest trip duration, we classified the months of the years into different seasons; winter,spring,summer and autumn. Because as a business consultant we may want to advice the company executives on possible planning for maintenace, repairs, renovations based on this output.

We made a data frame of the Month and due to the fact that we observed that the dataFrame was stored as a float we had each of the value of trip duration to integers and then map the month column to the season it falls in.

We then grouped by the seasons and made an aggreagation for the mean, sum, standard deviation and count of the trip duration

```python
df = data['Month']
Df = [int(i) for i in df]
DF = pd.DataFrame(Df)
seasonDur = DF[0]
#Create a Seasonal Dictionary that will map months to seasons
SeasonDict = {11:'Winter',12:'Winter',1:'Winter',2:'Spring',3:'Spring',4:'Spring',5:'Summer',6:'Summer',7:'Summer',
8:'Autumn',9:'Autumn',10: 'Autumn'}
mont = seasonDur.map(lambda x:SeasonDict[x])
tripDur = data['DurationMins']

seasonDf = pd.DataFrame(
    {'season':mont,
     'DurationMin':tripDur})
seasonDf
# Call the function with the groupby operation. 
Grouped = seasonDf.groupby(['season'])
Grouped.agg(['mean','sum', 'std','count'])
```

We observed from our output that the __summer__ season had the highest mean, sum, count trip duration. This makes so much sense because the weather conditions and temperature during this season is most favorable for such an outdoor sport like riding a bicycle.

### Problem 9

__What is the statictic of membershipType used by riders?__

We have different membership types and different menbership cards for whatever membership status a user holds <img src = https://cdn.styleblueprint.com/wp-content/uploads/2015/07/SB-Nash-bike-share-instructions-copy-800x600.jpg title= "membership card" width = "250" height = "200"> we would be interested to explore the membership types used by riders in this question.

We observed that some of the membership type was stored in different similar names by the users like "Annual, annual, Annual Membership Type, annual Membership", so we used regular expressions to help use Annual to represent all those with similar traits. The same applies for 24-Hour and 24-Hour kiosk, and then got the counts as well as made the plot

```python
import re
membershipTypes=data['MembershipType']
membershipTypeList=list(membershipTypes)
for i in range(len(membershipTypeList)):
    if re.search('(A|a)nnual',membershipTypeList[i]):
        membershipTypeList[i] = "Annual"
    elif re.search('24-Hour', membershipTypeList[i]):
        membershipTypeList[i] = "24-Hour"

membershipTypesdf = pd.DataFrame(membershipTypeList, columns = ['MembershipType'])
membershipTypesDF = pd.value_counts(membershipTypesdf['MembershipType'].values, sort=True)
membershipTypesDF        
```

```python
membershipTypesDF.plot(kind = "barh", title = "MembershipType")
plt.show()      
```

We observe that the 24-Hour kiosk is the membership type with the highest count, this makes sense because we know that most people don't like to only ride bicycle when they feel like or their friends want to, so they would prefer to only get a 24-hour membership ticket and then ride.

__What is the average carbon offset?__

Carbon offsetting means compensating for the carbon-dioxide pollution being made by preventing the same amount of pollution from happening elsewhere, In a surging rise on matters concerning global warming, and scientist tells us the world is slowly warming mostly because of the carbon-based, fossil fuels being burnt through industralization. In a bid to developing a clean green environment safe for habitation, industries are using Carbon offset to make a difference.

In theory, you could offset your carbon dioxide emmissions(also called green house emissions) by doing the environment good as well as yourself.

We simply calculated the average carbon offset by using the mean function on the estimated carbon offset data in our data and then plotted the histogram of the mean

```python
carbonOffset=data['EstimatedCarbonOffset']
print("The average estimated carbonOffset is ",carbonOffset.mean(), "calories per minute")
plt.hist(carbonOffset, range = (0,30),bins = 16)
plt.title('Average carbon offset')
plt.show()
```
